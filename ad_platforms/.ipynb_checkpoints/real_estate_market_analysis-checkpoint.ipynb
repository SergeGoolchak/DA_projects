Посчитаем процентное соотношение заполненных значений в столбцах к общему количеству строк.

```python
def not_nan_ratio(dataframe):
    ratio_df = dataframe.T.count(axis=1)
    ratio_df = pd.DataFrame(data=ratio_df, index=ratio_df.index).reset_index()
    ratio_df.columns=['columns_name', 'count_not_nan_values']
    ratio_df['ratio'] = ratio_df['count_not_nan_values'].map(lambda x: '{:.1%}'.format(x / 23699))
    return ratio_df
not_nan_ratio(data)
```

## Предобработка данных

### Работа с пропусками

Заполним пропуски в столбцах `floors_total`, `days_exposition` и `is_apartment` значениями 0.

```python
for column in ['floors_total', 'days_exposition', 'is_apartment']:
    data[column] = data[column].fillna(0)
```

Проверим заполнение пропущенных значений:

```python
data[['floors_total', 'days_exposition', 'is_apartment']].head(10)
```

Удалим строки с пропущенными значениями в столбце `locality_name`. Удаление не исказит наши данные, а вот наличие наименования населенного пункта критично важно для нашего исследования.

```python
data = data[data['locality_name'].isna() != True].reset_index(drop=True)
```

### Преобразование типов данных

Напишем функцию для удобного преобразования типов данных через циклы

```python
def change_type(dataframe, column: str, new_type: str):
    dataframe[column] = dataframe[column].astype(new_type)
    return dataframe
```

Изучим максимальные значения в столбцах `total_images`, `rooms`, `floors_total`, `floor`, `days_exposition`

```python
for column in [
    'total_images', 'rooms', 'floors_total', 
    'floor', 'days_exposition', 'total_area', 'last_price'
]:
    print(f'Максимальное значение в столбце {column}: {data[column].max()}')
```

Так как в столбцах `total_images`, `rooms`, `floors_total`, `floor` максимальное значение не больше 127, с помощью цикла `for` и написанной функции `change_type` преобразуем тип данных `int64` на `int8` в целях экономии памяти.

```python
for column in [
    'total_images', 'rooms', 'floors_total', 'floor'
]:
    change_type(data, column, 'int8')
```

С той же целью заменим тип данных в столбце `days_exposition` и `total_area` на `int16`

```python
for column in ['days_exposition', 'total_area']:
    change_type(data, column, 'int16')
```

В связи с тем что в колонке `last_price` значения измеряются в миллионах, заменим тип данных на `int32`, что также является более экономным вариантом в отличии от `int64`

```python
%%capture
change_type(data, 'last_price', 'int32')
```

Обработаем типы данных с оставшимися числовыми столбцами с типами `float64` и имеющими пропуски, заменим на  `float32`

```python
for column in [
    'ceiling_height', 'living_area', 'kitchen_area', 'balcony', 'airports_nearest', 
    'cityCenters_nearest', 'parks_around3000', 'parks_nearest', 'ponds_around3000', 'ponds_nearest'
]:
    change_type(data, column, 'float32')
```

В столбце `is_apartment` заменим тип даныых на булевые значения `bool`

```python
%%capture
change_type(data, 'is_apartment', 'bool')
```

`first_day_exposition` хранит в себе дату и время в формате строки, преобразуем тип данных в `datetime` и сохраним в столбце данные только о дате

```python
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format='%Y-%m-%d')
```

### Выведение дополнительных парамметров

В столбце `locality_name` мы имеем наименование населенного пункта и его тип (село, поселок, деревня и тд.). Мы планируем выделить тип населенного пункта в отдельный столбец для удобства анализа данных, а в столбце `locality_name` оставить только название населенного пункта.

Напишем функцию для выделения типа населенного пункта в `locality_name`

```python
def words_in_title(title):
    location_type = ' '.join(word for word in title.split(' ') if word.islower())
    return location_type
```

Затем функцию для выделения отдельно от типа, наименования населенного пункта

```python
def new_title(title):
    location_title = ' '.join(word for word in title.split(' ') if word.istitle())
    return location_title
```

Создадим новую колонку `location_type` куда перенесем типы населенных пунктов

```python
data.insert(14, 'location_type', data['locality_name'].fillna(' ').apply(words_in_title))
```

В связи с тем, что в столбце `locality_name` для городов не указывался их тип населенного пункта, мы добавим его вручную в `location_type`

```python
data.loc[(data.location_type == ''), 'location_type'] = 'город'
```

Перезапишем значения в колонке `locality_name`, оставим только названия населенных пунктов.

```python
data['locality_name'] = data['locality_name'].apply(new_title)
```

## Расчёты и добавление результатов в таблицу

### Расчет стоимости квадратного метра

Создадим новую колонку `cost_m2` и заполним ее расчитанными данными по цене за квадратный метр квартиры

```python
data.insert(3, 'cost_m2', (data['last_price'] / data['total_area']).round())

#преобразуем в нужный тип данных
data['cost_m2'] = data['cost_m2'].astype('float32')
```

### Создание новых столбцов с днем недели, месяцем и годом размещения объявления

Вычленим из дат записанных в столбце `first_day_exposition` необходиммые данные и запишем их новые колонки:

День недели в колонку `day`

```python
data.insert(5, 'day', data['first_day_exposition'].dt.day_name())
```

Месяц в колонку `month`

```python
data.insert(6, 'month', data['first_day_exposition'].dt.month)

#преобразуем в нужный тип данных
data['month'] = data['month'].astype('int8')
```

Год в столбец `year`

```python
data.insert(7, 'year', data['first_day_exposition'].dt.year)

#преобразуем в нужный тип данных
data['year'] = data['year'].astype('int16')
```

### Категоризация данных по этажу

Изучим данные об этажах в наших объявлениях, разобьем их на три категории "первый" - если это первый этаж, "последний" - если квартира находится на последнем этаже дома, все остальные этажи запишем в категорию "другое". Для категорий по этажам создадим новый столбец `floors_category`

```python
# Создаем новый столбец, и записываем в него категории 'последний' и 'другой' по квартирам с соответсвующими условиями
data['floors_category'] = np.where(data['floor'] == data['floors_total'], 'последний', 'другой')

# Вычисляем квартиры на первых этажах и добавляем для них категорию 'первый'
for i in data[data['floor'] == 1]['floors_category'].index:
    data.loc[i, 'floors_category'] = 'первый'
```

### Вычисление соотношения площадей жилой и кухни к общей

Расчитает соотношение площади кухни к общей площади квартиры и запишем данные в столбец `kt_area_ratio`

```python
data.insert(13, 'kt_area_ratio', (data['kitchen_area'] / data['total_area'] * 100))

#преобразуем в нужный тип данных
data['kt_area_ratio'] = data['kt_area_ratio'].astype('float32')
```

Повторим данную операцию для жилой площади к общей

```python
data.insert(12, 'lt_area_ratio', (data['living_area'] / data['total_area'] * 100))

#преобразуем в нужный тип данных
data['lt_area_ratio'] = data['lt_area_ratio'].astype('float32')
```

### Изучение результатов проделанныч расчетов

Выведем на экран наши новые колонки

```python
data[
    ['cost_m2', 'day', 'month', 'year', 'total_area', 
     'living_area', 'lt_area_ratio', 'kitchen_area', 'kt_area_ratio']
]
```

```python
data.info()
```

### Изучение параметров площадь, цена, число комнат и высота потолков

Проставим признак True по квартирам с количеством комнат 0 и малой площадью.

```python
for i in data[(data['rooms'] == 0) & (data['total_area'] < 40)].index:
    data.loc[i, 'studio'] = True
```

и выведем на экран статистическое описание параметра

```python
data_filtered = data_filtered[
    (data_filtered['ceiling_height'] > 2) & (data_filtered['ceiling_height'] <= 10)
]
```

```python
data_filtered = data_filtered[data_filtered['days_exposition'] <= 500]
data_filtered['days_exposition'].hist(bins=100, figsize=(15, 5))
plt.title('Гистограмма распределения значений параметра days_exposition')
plt.xlabel('Количество дней с даты размещения объявления до закрытия')
plt.show()
```

После того как мы отбросили значения которые не входили в "ящик с усами", мы обнаружили, что достаточно большое количество продаж в значениях от 0 до 5 дней, это выглядит неправдоподобным, скорее всего люди просто снимали объявление передумав продавать, поэтому мы избавимся от этих значений.
Также значения в районе 45, 60 и 90 дней выделяются своей частотой. Скорее всего это связано с условиями размещения на площадке. Уберем эти значения и заново изучим данные по продаже.

Изучим отфильтрованные данные, проистроив гистограмму

```python
data_filtered['days_exposition'].hist(bins=100, figsize=(15, 5))
plt.title('Гистограмма распределения отфильтрованных значений параметра days_exposition')
plt.xlabel('Количество дней с даты размещения объявления до закрытия')
plt.show()
```

Отлично! Теперь выведем описание данных

```python
data_filtered['days_exposition'].describe()
```

После изменения нашей выборки, мы выявили, что средняя продолжительность продажи квартиры занимает от 40 до 199 дней.

### Исследование рынка квартир Санкт-Петербурга

```python
spb_price['cityCenters_nearest_km'].corr(spb_price['last_price']).round(3)
```

В случае для Санкт-Петербурга мы отчетливо видим сильную отрицательную зависимость. Чем меньше расстояние до центра города, тем дороже цена на жилье. Однако по удалению от центра есть также участки с дорогой недвижимостью, так называемые "элитные" районы.<br><br>

- `total_images` — число фотографий квартиры в объявлении
- `last_price` — цена на момент снятия с публикации
- `total_area` — площадь квартиры в квадратных метрах
- `cost_m2` - цена квадратного метра в рублях
- `first_day_exposition` — дата публикации
- `day` — день недели даты публикации объявления
- `month` — месяц даты публикации объявления
- `year` — год даты публикации объявления
- `rooms` — число комнат
- `ceiling_height` — высота потолков в метрах
- `floors_total` — всего этажей в доме
- `living_area` — жилая площадь в квадратных метрах
- `lt_area_ratio` — соотношение жилой площади к общей
- `floor` — этаж
- `kt_area_ratio` — соотношение площади кухни к общей площади
- `is_apartment`  — апартаменты(да или нет)
- `studio` — квартира-студия (да или нет)
- `open_plan` — свободная планировка
- `kitchen_area` — площадь кухни в квадратных метрах
- `balcony` — число балконов
- `location_type` — тип населённого пункта
- `locality_name` — название населённого пункта
- `airports_nearest` — расстояние до ближайшего аэропорта в метрах
- `cityCenters_nearest` — расстояние до центра города в метрах
- `parks_around3000` — число парков в радиусе 3 км
- `parks_nearest` — расстояние до ближайшего парка в метрах
- `ponds_around3000` — число водоёмов в радиусе 3 км
- `ponds_nearest` — расстояние до ближайшего водоёма в метрах
- `days_exposition` — сколько дней было размещено объявление(от публикации до снятия)
- `floors_category` — категория этажа(первый, другой, последний)

В архиве объявлений было большое количество пропущенных значений, рекомендуется проанализировать причины подобных пропусков и найти пути их предупреждения. Например в анкете объявления сделать все поля обязательными к заполнению, а в части из них не заполнение, а после с выбором вариантов из предложенных, как в случае со студиями или аппартаментами. <br>
Для работы с полученными данными, нам пришлось провести их предобработку. Мы заполнили часть пропущенных значений, а какую-то часть строк с пропущенными значениями и вовсе удалили. Преобразовали типы данных практически в каждой колонке, что дало нам экономию, с учетом добавления в датасет новой информации, мы сэкономили 33% памяти.<br>
Для проведения исследования рынка недвижимости, мы подготовили колонки с новыми данными:<br>
       - Расчитали цену за квадратный метр по каждой квартире, а также соотношения площадей жилой и кухни к общей площади квартиры. Все расчеты записали в новые столбцы датафрема.<br>
       - Разбили дату на составные части (день недели, месяц, год) и добавили в соответсвующие колонки датасета.<br>
       - Создали категории для расположния квартиры на этаже, у нас их три типа: первый, другой и последний.<br><br>
Исследование мы проводили изучая объявления как по всей выборке, так и отдельно по Санкт-Петербургу. Мы выяснили, что:
 1. Время продажи квартиры в среднем занимает от 40 до 199 дней.
 2. Изучили, зависит ли цена от площади квартиры, числа комнат, этажа, удалённости от центра, а также от даты размещения: дня недели, месяца и года. Мы выявили цена больше всего зависит от площади квартиры и близости к центру города, квартиры на последнем этаже оцениваются дороже квартир на первых этажах, а также от года ее продажи.
 3. Отобрали 10 населённых пунктов с наибольшим числом объявлений, ими оказались:
    - Санкт-Петербург
    - Мурино
    - Кудрово
    - Шушары
    - Всеволожск
    - Пушкин
    - Колпино
    - Парголово
    - Гатчина
    - Выборг<br>
    И выявили, что самая лорогая средняя цена за квадтратный метр в Зеленогорске, а самая низкая - в Старополье.
 4. Построили график для квартир Санкт-Петербурга, по средней стоимости жилья для каждого километра отдалености от центра города. Отчетливо увидели сильную отрицательную зависимость. Чем меньше расстояние до центра города, тем дороже цена на жилье. Однако по удалению от центра есть также участки с дорогой недвижимостью, так называемые "элитные" районы. Изучив данный график, определили, что центром считается радиус в районе 3 километров.
 5. Построили матрицу диаграмм по столбцам: площадь, цена, число комнат, высота потолков. Также построили графики взаимосвязи стоимости  квартиры с параметрами: число комнат, этаж, удалённость от центра, дата размещения объявления. Выяснили, что в центре города квартиры преобладают с потолками в 3 метра, тогда как в общей выборке от 2.5 от 2.8 метров. В центральном радиусе Санкт-Петербурга расстояние до центра уже практически не влияет на цену. Средняя стоимость на квартиры в центре выше, а также преобладают многокомнатные квартиры.
