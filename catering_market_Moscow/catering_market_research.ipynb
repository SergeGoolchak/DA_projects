### Изучение локаций размещения заведений общественного питания в Москве

Для анализа локаций размещения заведений нам потребутся выделенная информация об улицах из адресе заведения, для последующей группировке по ним. Напишем функцию вычлиняющую улицу из столбца адрес.

```python
def street_from_address(df, column: str, new_column_name: str):
    list_streets = []
    for value in data[column]:
        if len(value.split(',')) < 3:
            list_streets.append(value.split(',')[0])
        else:
            list_streets.append(value.split(',')[1])
    df[new_column_name] = list_streets
    df[new_column_name] = df[new_column_name].str.lstrip()
    return df
```
